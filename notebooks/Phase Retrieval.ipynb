````python
npix = 257
PSF_measured = test_data.copy()
nz, ny, nx = PSF_measured.shape;
zc, yc, xc = unravel_index(PSF_measured.argmax(), PSF_measured.shape);
PSF_offset = 1.0 * bincount(PSF_measured.ravel()).argmax()
PSF = zeros((nz, npix, npix))
PSF[:, npix // 2 - yc:npix // 2 + ny - yc, npix // 2-xc:npix // 2 + nx - xc] = PSF_measured
PSF = sqrt(PSF)
PSF = PSF - sqrt(PSF_offset)
PSF *= PSF > 0
PSF_offset
````

````python
def retrieve_phase(data, psf, pupil=None, normed=True):
    """This function acurately reproduces the matlab code"""
    if pupil is None:
        pupil = psf._gen_pupil()
    psf._gen_psf(pupil)
    phase = angle(psf.PSFa.squeeze())
    new_psf = data * exp(1j * phase)
    new_pupils = fftn(fftshift(new_psf, axes=(1,2)), axes=(1,2))
    new_pupil = (new_pupils / psf._calc_defocus()).mean(0)
    if normed:
        new_pupil = exp(1j * angle(new_pupil))
    return psf._gen_pupil() * new_pupil, new_pupils # why do we do this?
````

````python
from scipy.ndimage import label, labeled_comprehension
````

````python
def floodfill(img, seed):
    """Given seed return filled image points"""
    img2 = logical_not(img)
    labelled_img, num_feat = label(img2)
    seed_val = labelled_img[seed]
    return labelled_img == seed_val
````

````python
def PhaseResidues(IM_phase):
    """
    PhaseResidues.m calculates the phase residues for a given wrapped phase
    image. Note that by convention the positions of the phase residues are 
    marked on the top left corner of the 2 by 2 regions.
    
      active---res4---right
         |              |
        res1           res3
         |              |
      below---res2---belowright
    Phase residues with integer multiples of 2*pi are not accounted for, but 
    these rarely occur.
    Created by B.S. Spottiswoode on 07/10/2004
    Last modified on 08/10/2004
    """
    #The code below is simply a vectorised representation of operations on 2 by 2
    #blocks in the matrix
    IM_active = IM_phase.copy()
    IM_below = zeros_like(IM_phase)
    IM_below[:-1,:] = IM_phase[1:,:]  # Shift up 1 pixel
    IM_right = zeros_like(IM_phase)
    IM_right[:,:-1] = IM_phase[:,1:]  # Shift left 1 pixel
    IM_belowright = zeros_like(IM_phase)
    IM_belowright[:-1,:-1] = IM_phase[1:,1:]  # Shift up 1 pixel & left 1 pixel

    # Wrap the phase differences as we loop around the 2 by 2 blocks
    res1 = mod(IM_active - IM_below + pi, 2 * pi) - pi      # Up - Down (bottom row is unreliable)
    res2 = mod(IM_below - IM_belowright + pi, 2 * pi) - pi  # Left- Right, shifted up (bottom row & far-right column are unreliable)
    res3 = mod(IM_belowright - IM_right + pi, 2 * pi) - pi  # Down - Up, shifted left (bottom row & far-right column are unreliable)
    res4 = mod(IM_right - IM_active + pi, 2 * pi) - pi      # Right - Left, (far-right column is unreliable)

    temp_residues = res1 + res2 + res3 + res4              #Sum the phase differences. Positive residues appear as 2*pi, negative as -2*pi.
    # Far-right column and bottom row are unreliable in temp_residues
    residues = (temp_residues >= 6).astype(int)                    #Assign 1 to positive residue (which should equal 2*pi)
    residues -= (temp_residues <= -6).astype(int)        #Assign -1 to negative residues (which should equal -2*pi)
    residues[:,-1] = 0
    residues[-1,:] = 0         #Zero pad the border residues
    return residues
````
